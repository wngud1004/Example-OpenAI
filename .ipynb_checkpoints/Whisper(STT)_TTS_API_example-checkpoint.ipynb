## API 키 로딩
API 키가 들어간 token.txt를 설정해줘야 한다.

혼자 쓰는 노트북이면 그냥 os.environ["OPENAI_API_KEY"] = "키" 해도 된다. 
주의: 최대 20불 리밋 걸어두지 않으면 돈이 계속 빠져나갈수도..

```python
from openai import OpenAI
import os

with open('token.txt', 'r') as file:
    os.environ["OPENAI_API_KEY"] = file.read().strip()

openai.api_key = os.getenv("OPENAI_API_KEY")
```

오디오 파일 오픈하고 그냥 transcription 콜 하면 사용 가능하다.

```python
client = OpenAI()

audio_file= open("intro-draft.m4a", "rb")
transcription = client.audio.transcriptions.create(
  model="whisper-1",
  file=audio_file
)
print("Done")
```

```python
# 내용
print(transcription.text)
print("")
```

## 오디오 파일 만들기

```python
speech_file_path = "speech_output.mp3"
response = client.audio.speech.create(
  model="tts-1",
  voice="alloy",
  input="안녕하세요 얼굴이 동그래서 슬픈 양파입니다"
)
response.stream_to_file(speech_file_path)
```

```python
from IPython.display import Audio, display
Audio("speech_output.mp3")  # From URL
```

```python
audio_file= open("intro-draft.m4a", "rb")
translation = client.audio.translations.create(
  model="whisper-1",
  file=audio_file
)
print(translation.text)
```
